```python
import matplotlib.pyplot as plt
import matplotlib.image as img 
import os

current_path = os.getcwd()

combinations = ["additive", "multiplicative1", "multiplicative2"]

# dict_d = {0.0: "d_0", 0.5 : "d_05", 1.0 : "d_1"}
# ds = [0.0, 0.5, 1.0]

dict_d = {1.0 : "d_1"}
ds = [1.0]

distributions = ["LogNormal", "Gamma"]

series = ["ena", "carga"]#, "airline"]

colunas = ["Combinacao", "d", "Distribuicao", "Serie", "MAPE Treino", "MAPE Teste"]
```

```python
dict_imagens = dict()
```

```python
plot_types = ["fit_in_sample", "forecast", "components", "forecast_histograms", 
            "quantile_qqplot", "quantile_residuals_acf", "quantile_residuals", "quantile_residuals_histogram"]

for serie in series:
    dict_imagens[serie] = dict()
    for combination in combinations:
        dict_imagens[serie][combination] = dict()
        for distribution in distributions:
            dict_imagens[serie][combination][distribution] = dict()
            for d in ds:
                dict_imagens[serie][combination][distribution][d] = dict()
                dir_plot = current_path+"\\Saidas\\CombNaoLinear\\LinkFunction\\"+combination+"\\"+dict_d[d]+"\\"+distribution+"\\"
                list_files = os.listdir(dir_plot)
                for plot_type in plot_types:
                    name_file = serie+"_"+plot_type+"_"+distribution+".png"
                    if  name_file in list_files:
                        img_ = img.imread(dir_plot+name_file)
                        dict_imagens[serie][combination][distribution][d][plot_type] = img_
                
```

```python

dict_titles = {"fit_in_sample": "Fit in Sample",
            "forecast": "Forecast",
            "quantile_residuals_acf": "FAC Quantile Residuals", 
            "quantile_residuals": "Quantile Residuals", 
            "quantile_qqplot": "QQPlot Quantile Residuals" , 
            "quantile_residuals_histogram": "Histogram Quantile Residuals", 
            "forecast_histograms":"Forecast Quantiles",
            "components": "Components"}

dict_names = {"fit_in_sample": "Fit in Sample_",
            "forecast": "Forecast_",
            "quantile_residuals_acf": "FAC Quantile Residuals_", 
            "quantile_residuals": "Quantile Residuals_", 
            "quantile_qqplot": "QQPlot Quantile Residuals_" , 
            "quantile_residuals_histogram": "Histogram Quantile Residuals_", 
            "forecast_histograms":"Forecast Quantiles_",
            "components" : "Components"}
```

```python

for serie in series:
    print(serie)
    for combination in combinations:
        print(" %s"%combination)
        for plot_type in plot_types:
            print("     %s"%plot_type)            
            fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(7,4), constrained_layout =True) #10,10 para relatorio
            fig.suptitle("%s - %s - %s"%(dict_titles[plot_type],serie,combination))
            # fig.tight_layout()
            # fig.subplots_adjust(left=0.1,
            #         bottom=0.1, 
            #         right=0.2, 
            #         top=0.2, 
            #         wspace=0.2, 
            #         hspace=0.2)

            ax[0,0].imshow(dict_imagens[serie][combination]["LogNormal"][0.0][plot_type])
            ax[0,1].imshow(dict_imagens[serie][combination]["LogNormal"][0.5][plot_type])
            ax[0,2].imshow(dict_imagens[serie][combination]["LogNormal"][1.0][plot_type])    
            ax[1,0].imshow(dict_imagens[serie][combination]["Gamma"][0.0][plot_type])
            ax[1,1].imshow(dict_imagens[serie][combination]["Gamma"][0.5][plot_type])
            ax[1,2].imshow(dict_imagens[serie][combination]["Gamma"][1.0][plot_type])  
            ax[0,0].axis("off")
            ax[0,1].axis("off")
            ax[0,2].axis("off")
            ax[1,0].axis("off")
            ax[1,1].axis("off")
            ax[1,2].axis("off")
            # plt.savefig("Saidas\\Relatorio\\%s_%s_%s.jpg"%(dict_names[plot_type],serie, combination))
            plt.savefig("Saidas\\Slides\\%s_%s_%s.jpg"%(dict_names[plot_type],serie, combination),dpi=300)
```
